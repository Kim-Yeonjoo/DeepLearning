```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import tensorflow_datasets as tfds
train_data, validation_data, test_data = tfds.load(
    name='imdb_reviews',
    split=('train[:60%]', 'train[60%:]', 'test'),
    as_supervised=True)
```

```python
from transformers import pipeline
```

```python
# 모델 이용해서 분류기 생성
classifier = pipeline('sentiment-analysis', model='distilbert-base-uncased-finetuned-sst-2-english')
```

```python
# 분류기에 문장 넣기
sent = ['Today is a beautiful day!', 'we all still have terrible days']

result = classifier(sent, truncation=True)  # 입력이 길면 자르는 옵션
result
```

```python
cnt = 1000

sentences, labels = next(iter(train_data.batch(cnt)))
```

```python
sent_bytes = sentences.numpy().tolist()

# 클렌징
sent_list = []
for ss in sent_bytes:
    sent_list.append(ss.decode('utf-8'))    # 텍스트 변환
sent_list[:3]
```

```python
# 분류기에 넣기
result = classifier(sent_list, truncation=True)
```
