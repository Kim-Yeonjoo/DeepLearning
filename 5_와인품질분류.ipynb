# 신경망 생성 - 와인등급 데이터

```python
import pandas as pd

redwine = pd.read_csv('winequality-red.csv', sep=';')
redwine.head()
```

```python
from sklearn.model_selection import train_test_split

redwine_X = redwine.iloc[:, :-1].to_numpy()
# 원핫인코딩
redwine_y = pd.get_dummies(redwine.iloc[:, -1]).to_numpy()

# train / test
train_X, test_X, train_y, test_y = train_test_split(redwine_X, redwine_y, test_size = 0.3, random_state=1)
train_X.shape, train_y.shape
```

# 문제 정의: 와인의 품질속성을 분류하기

1. 모형 설계
 - Sequential()
 - 레이어 1 : 뉴런 50, 활성화함수 relu
 - 레이어 2 : 뉴런 50, 활성화함수 relu
 - 레이어 3 : 뉴런 30, 활성화함수 relu
 - 레이어 4 : 출력 6, 활성화함수 softmax (분류, 확률계산)

2. 속성
 - compile()
 - 최적화: adam
 - 손실함수: categorical_crossentropy
 - 평가지표: accuracy

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input

#1. 모델정의
model = Sequential()
model.add(Input(shape=(11,)))  # 입력층 11개 피처 지정
model.add(Dense(50, activation='relu'))   # 은닉층 레이어1
model.add(Dense(50, activation='relu'))   # 레이어2
model.add(Dense(30, activation='relu'))   # 레이어3
model.add(Dense(6, activation='softmax'))  # 출력층 레이어4

model
```

```python
#2. 속성 설정 - 컴파일
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
#3. 모델 훈련
model.fit(train_X, train_y, epochs=200, batch_size=50, verbose=1)
```

```python
#4. 평가
pred = model.predict(test_X)
pred
```

```python
#4-1. 제일 높은 평가 출력
import numpy as np
pred_ag = np.argmax(model.predict(test_X), axis=1)
pred_ag
```

```python
#5. 실제값과 비교
np.argmax(test_y, axis=1)
```

```python
pd.crosstab(np.argmax(test_y, axis=1), pred_ag, margins=True)
```

```python
#6. 수치로 확인하기
loss, accuracy = model.evaluate(test_X, test_y)
print('loss : ', loss)
print('accuracy : ', accuracy)
```

# 정확도 올리기
1. 모형 구조 변경
2. 최적화 함수 바꾸기
3. 에포크 횟수 늘리기

```python
# 모형 구조 재설정
model_1 = Sequential()
model_1.add(Input(shape=(11,)))  # 입력층 11개 피처 지정
model_1.add(Dense(128, activation='relu'))   # 은닉층 레이어1
model_1.add(Dense(128, activation='relu'))   # 레이어2
model_1.add(Dense(64, activation='relu'))   # 레이어3
model_1.add(Dense(6, activation='softmax'))  # 출력층 레이어4

model_1.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# epochs 횟수 늘리기
model_1.fit(train_X, train_y, epochs=500, batch_size=50, verbose=1)
```

```python
loss, accuracy = model_1.evaluate(test_X, test_y)
print('loss : ', loss)
print('accuracy : ', accuracy)
```

```python
model_2 = Sequential()
model_2.add(Input(shape=(11,)))  # 입력층 11개 피처 지정
model_2.add(Dense(164, activation='relu'))   # 은닉층 레이어1
model_2.add(Dense(164, activation='relu'))   # 레이어2
model_2.add(Dense(128, activation='relu'))   # 레이어3
model_2.add(Dense(6, activation='softmax'))  # 출력층 레이어4

model_2.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# epochs 횟수 늘리기
model_2.fit(train_X, train_y, epochs=300, batch_size=50, verbose=1)
```

```python
loss, accuracy = model_2.evaluate(test_X, test_y)
print('loss : ', loss)
print('accuracy : ', accuracy)
```
